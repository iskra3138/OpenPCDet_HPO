```python
import os
import glob

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Modify the belows according to your paths

TRAIN_LABEL_FILE = '/nas2/YJ/git/tree_data_raw/pcdet/data/tree/ImageSets/train.txt'
VALID_LABEL_FILE = '/nas2/YJ/git/tree_data_raw/pcdet/data/tree/ImageSets/val.txt'
TEST_LABEL_FILE = '/nas2/YJ/git/tree_data_raw/pcdet/data/tree/ImageSets/test.txt'

TRAIN_LABEL_ROOT = '/nas2/YJ/git/tree_data_raw/pcdet/data/tree/training/labels/'
VALID_LABEL_ROOT = '/nas2/YJ/git/tree_data_raw/pcdet/data/tree/training/labels/'
TEST_LABEL_ROOT = '/nas2/YJ/git/tree_data_raw/pcdet/data/tree/testing/labels/'
```

```python
truncations = []
occlusions = []
lengths = []
widths = []
heights = []
```

### Read Train Files

```python
num_file = 0
num_instance = 0

with open(TRAIN_LABEL_FILE, 'rb') as f : 
    file_nums = f.readlines()
    for file_num in file_nums :
        label_file = os.path.join(TRAIN_LABEL_ROOT, file_num.strip().decode("utf-8") + '.txt')
        with open(label_file, 'r') as f:
            num_file += 1
            lines = f.readlines()
            for line in lines:
                label = line.strip().split(' ')
                t = float(label[1])
                o = float(label[2])
                h = float(label[8])
                l = float(label[9])
                w = float(label[10])
                
                num_instance += 1
                
                truncations.append(t)
                occlusions.append(o)
                heights.append(h)
                lengths.append(l)
                widths.append(w)

print (num_file, num_instance)
```
